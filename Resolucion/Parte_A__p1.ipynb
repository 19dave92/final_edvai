## 2.2) Desglose de la columna customerEmail

Tenemos columnas que podrian contener información agrupada como es el caso de 'customerEmail' donde detemos 'direccion@proveedor.dominio'. Proveedor y dominio son datos de interes que pueden revelarnos patrones en los datos, para ello debemos descomponer la información y acomodarlas en columnas que sean mas descritpivas

````python
data_copy["customerEmail"]
````

````python
dominios_email= []   # '.com', '.cl', '.info', ...
proveedor_email= []  # proveedor: yahoo, hotmail, gmail...
proveedores_populares = ['yahoo', 'gmail', 'hotmail']

for email in data_copy['customerEmail']:
    try:
        """
        Si email = "edvai@yahoo.com"

        aux = yahoo.com
        mail = yahoo
        domain = com
        """
        # Split devuelve una lista
        aux = email.split('@')[1] #Separamos el email en dos, nos quedamos con el segundo elemento que es el correo.com
        dominio = aux.split('.')[1] #Separamos el dominio en dos, y nos quedamos con el elemento despues del punto
        proveedor = aux.split('.')[0] #Guardamos el elemento antes del punto
        
        # Verificamos que el mail se encuentre en la lista de proveedores regulares
        if proveedor in proveedores_populares:
            proveedor_email.append(proveedor)
        else:
            proveedor_email.append('other')
        dominios_email.append(dominio)
    # La excepcion se levanta cuando el formato de mail no es convencional o hay caracteres desconocidos.    
    except:
        dominios_email.append('weird')
        proveedor_email.append('weird')
        # print(email)
````

Agregamos las dos listas al dataframe

````python
data_copy['emailProvider'] = proveedor_email
data_copy['emailDomain'] = dominios_email
````

Como ya fue tratada la variable customerEmail, ya no la necesitamos

````python
data_copy = data_copy.drop(['customerEmail'], axis=1)
````

### 2.3) "Categorización" de la variable 'customerIPAdress'

Observamos que las IP a las que estamos habituados tiene una longitud de 15 caracteres (incluyendo el punto).
```
longitudIP = len(255.255.255.255)
print(longitudIP)
# 15
```

Vamos a separa las IP en dos categorias: de 15 caracteres y mayor a 15:

__**NOTA**__: Decidimos NO CONSERVAR la IP de forma indiviual porque ocasionaría que tengamos tantos valores unicos como cantidad de registros. Esto es un inconveniente debido a que a la hora de hacer AED es mas facil tratar variables que tengan la menor cantidad de datos unicos. 

Si guardamos y contemplamos la IP en su forma original tendremos N valores unicos donde N es la cantidad de registros (debido a que la IP es unica para cada caso).
Mientras que si tenemos dos categorias, posteriormente será mucho mas facil para el modelo procesar la variable.  

````python
direccion_IP = []
for direccion in data_copy['customerIPAddress']:

    # A veces lo toma como flotante al address, para evitar problemas, lo convertimos
    aux_address = str(direccion)
    
    if len(aux_address) > 15:
        direccion_IP.append('digits_and_letters')
    else:
        direccion_IP.append('only_letters')

direccion_IP
````

Agregamos la nueva variable al dataframe y eliminamos la columna customerIPAdress

````python
data_copy['customerIPAddressSimplified'] = direccion_IP

data_copy = data_copy.drop(['customerIPAddress'], axis=1)

data_copy.head()
````

## 2.4) Lugar desde donde de la compra VS direccion de envio *(deberian coincidir)*

**Creamos una función que recibe como parametro el nombre de la columna que queremos iterar. Tomamos el mismo criterio que usamos con customerEmail:

````python
def extraer_nombre_ciudad(name_column):
    lista_ciudades = []

    for direccion in data_copy[name_column]: #Iteramos cada elemento de la columna
        try:
            """
            Si address = "5493 Jones Islands\nBrownside, CA 51896"

            aux = "Brownside, CA 51896"
            ciudad_con_CP = "CA 51896"
            ciudad = "CA"
            """
            aux = direccion.split('\n')[1]
            ciudad_con_CP = aux.split(', ')[1]
            ciudad = ciudad_con_CP.split(' ')[0]
            
            lista_ciudades.append(ciudad) #Agregamos solamente el nombre de la ciudad
                
        except:
            # Si no aparece el nombre de la ciudad asignamos: unknown
            lista_ciudades.append("unknown")
    
    return lista_ciudades
````

Utilizaremos la funcion creada (extraer_nombre_ciudad) para "limpiar" las variables 
**customerBillingAddress** y **orderShippingAddress** y quedarnos solo con las ciudades en cada variable. De esta forma nos será más fácil poder compararlas entre si. 

Nota personal: supongo que luego nos quedaremos con la comparacion, agregaremos una columna que diga si origen y destino coinciden y eliminaremos las variables de que contienen las cidudades.

````python
data_copy["customerBillingAddress"] = extraer_nombre_ciudad('customerBillingAddress') #Solo enviamos el nombre de la columna
data_copy["orderShippingAddress"] = extraer_nombre_ciudad('orderShippingAddress')

data_copy.head()
````

Hacemos la comparacion de las ciudades:
* Si coinciden -> YES
* No coinciden -> NO
* Uno es desconocido -> Unknow

````python
misma_ciudad = []

for indice in range(len(data_copy)): # len(data_copy) nos da la cantidad de registros
    direccion_compra = data_copy['customerBillingAddress'][indice]
    direccion_envio = data_copy['orderShippingAddress'][indice]

    response = "unknown"

    if direccion_compra != "unknown" and direccion_envio != "unknown":
        if direccion_compra == direccion_envio:
            response = "yes"
        else:
            response = "no"
    
    misma_ciudad.append(response)
````

````python
status(data_copy)
````

Creamos una columna/variable nueva para agregar la informacion de las ciudades coincidentes al dataframe. Posteriormente eliminamos las columnas "customerBillingAddress" y "orderShippingAddress". Obtenemos los primeros cinco registros para corroborar los cambios.

````python
# Creamos la columna y la llenamos
data_copy["sameCity"] = misma_ciudad
# Eliminamos las columnas que ya no necesitamos
data_copy = data_copy.drop(["customerBillingAddress", "orderShippingAddress"], axis=1)
# Visualizamos los cambios
data_copy.head()
````
