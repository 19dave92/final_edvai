```python
import pandas as pd
import json

# Si el archivo es demasiado extenso, puedes cargarlo en un diccionario primero.
with open('../data/customers_2.json', 'r') as f:
    data = json.load(f)

# Luego, convierte esa lista de diccionarios a un DataFrame
df = pd.DataFrame(data)

# Muestra las primeras filas para verificar
df.head()
```

```python
from pandas import json_normalize

clientes_df = json_normalize(df['customer'])

clientes_df.head(5)
```
